```python
# Generating random emails and passwords for user login
import random
import string

emails = []
passwords = []

def email_unique():
    email = random_email()
    if email not in emails:
        return email
    else:
        email_unique()
    
for user in range(len(df)):
    email = email_unique()
    password = random_password()
    emails.append(email)
    passwords.append(password)

print(emails[0:5])
```
