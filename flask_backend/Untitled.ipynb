```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

labels={}
labels[1]=r'$a$'
labels[2]=r'$b$'
labels[3]=r'$c$'
labels[4]=r'$d$'
labels[5]=r'$e$'
labels[6]=r'$f$'
labels[7]=r'$g$'
labels[8]=r'$h$'
labels[9]=r'$i$'
labels[10]=r'$j$'

def draw(G, pos, measures, measure_name, labels):
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250, cmap=plt.cm.plasma, 
                                   node_color=list(measures.values()),
                                   nodelist=measures.keys())
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
    # labels = nx.draw_networkx_labels(G, pos)
    edges = nx.draw_networkx_edges(G, pos)   
    nx.draw_networkx_labels(G,pos,labels,font_size=16)

    plt.title(measure_name)
    plt.colorbar(nodes)
    plt.axis('off')
    plt.show()
    
DiG = nx.DiGraph()
DiG.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 6), (2, 4), (2, 5), (2, 7), (3, 4), (3, 6), (4, 5), 
                   (4, 6), (4, 7), (5, 7), (6, 7), (6, 8), (7, 8), (8, 9), (9, 10)])
dpos = {1: [0.1, 0.9], 2: [0.4, 0.8], 3: [0.8, 0.9], 4: [0.15, 0.55],
        5: [0.5,  0.5], 6: [0.8,  0.5], 7: [0.22, 0.3], 8: [0.30, 0.27],
        9: [0.38, 0.24], 10: [0.7,  0.3]}

NewDiG = DiG.to_undirected() 
draw(DiG, dpos, nx.in_degree_centrality(DiG), 'DiGraph Degree Centrality', labels)
```

```python
draw(NewDiG, dpos, nx.closeness_centrality(NewDiG), 'Closeness Centrality', labels)
print(labels)
print(nx.closeness_centrality(DiG))
```

```python
draw(NewDiG, dpos, nx.eigenvector_centrality(NewDiG), 'DiGraph Eigenvector Centrality', labels)
print(nx.eigenvector_centrality(NewDiG))
```

```python
draw(NewDiG, dpos, nx.betweenness_centrality(NewDiG), 'Betweenness Centrality', labels)
print(nx.betweenness_centrality(NewDiG))
```
